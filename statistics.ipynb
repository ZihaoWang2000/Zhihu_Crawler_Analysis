```python
import codecs
import jieba
import jieba.posseg as pseg
import datetime

def stop_key_num(text):
    # sentence_seged = jieba.cut(text)
    stopkey = [w.strip() for w in codecs.open('./stop_dic/stopwords.txt', 'r',encoding='utf-8').readlines()]
    seg_list = pseg.cut(text)  # 精确模式分词[默认模式]
    num = 0
    for word in seg_list:
        if word.word in stopkey:
            num = num + 1  # 分词追加到列表
    return num
```

```python
sql_1 = 'select answer_text, first_post_time, question_time from zhihu_car_cleaned'
data_1 = Data_admin.database(sql_1)
import pandas as pd
data = pd.DataFrame(data_1)
data
```

```python
import re
new_data = []
for index in data.index:
    text = data.loc[index].values[0]
    length = len(text)
    para = len(re.findall('[\\n]',text))+1
    number = len(re.findall(r'[-+]?[0-9]*\.?[0-9]+',text))
    stop_key = stop_key_num(text)
    answer_time = data.loc[index].values[1]
    question_time = data.loc[index].values[2]
    response_time = (answer_time - question_time).days
    new_data.append([index+1,text,length,para,number,stop_key,response_time])
```

```python
all = pd.DataFrame(new_data)
all.columns = ['id','text','length','para','number','stop_key','response_time']
all
```

```python
import tqdm

my_data = all[['id','length','para','number','stop_key','response_time']]
# print(my_data,my_data.loc[1][1])

for index in tqdm.tqdm(my_data.index):
    sql_2 = "update zhihu_car_cleaned set text_length={} , para_num = {},  number_num = {}, stop_key_num = {}, response_time = {} where answer_id = {}".format(int(my_data.loc[index].values[1]), int(my_data.loc[index].values[2]), int(my_data.loc[index].values[3]), int(my_data.loc[index].values[4]), int(my_data.loc[index].values[5]),my_data.loc[index].values[0])
    Data_admin.database(sql_2)
```
